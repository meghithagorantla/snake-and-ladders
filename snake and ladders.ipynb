```python

import random

numbers=100

   
def check_ladder(pts):
    if(pts==2):
        print('Ladder')
        return 23
    elif(pts==6):
        print('Ladder')
        return 45
    elif(pts==20):
        print('Ladder')
        return 59
    elif(pts==52):
        print('Ladder')
        return 72
    elif(pts==57):
        print('Ladder')
        return 96
    elif(pts==71):
        print('Ladder')
        return 92
   
    else:
        return pts
   
 
def check_snake(pts1):
    if(pts1==43):
        print('snake')
        return 17
    elif(pts1==50):
        print('snake')
        return 5
    elif(pts1==56):
        print('snake')
        return 8
   
    elif(pts1==73):
        print('snake')
        return 15
   
    elif(pts1==84):
        print('snake')
        return 58
    elif(pts1==87):
        print('snake')
        return 49
   
    elif(pts1==98):
        print('snake')
        return 40
    else:
        return pts1
   

def reached_numbers(pts):
    if pts==numbers:
        return True
    else:
        return False
       
   
def play():
   
     p1_name=input('player1, please enter ur name:')    
     p2_name=input('player2, please enter ur name:')    
     
     pp1=0
     pp2=0
     turn=0
     while(1):
         if(turn%2==0):
             print(p1_name,'your turn')
             
             c=input('press 1 to continue 0 to quit:')
             if(c==0):
                 print(p1_name,'scored',pp1)
                 print(p2_name,'scored',pp2)
                 print('Quitting the game..thanks for playing')
                 break
             
             dice=random.randint(1,6)  
             print('dice showed',dice)
             pp1=pp1+dice
             pp1=check_ladder(pp1)
             pp1=check_snake(pp1)
             if pp1>numbers:
                 pp1=numbers
             print(p1_name,"ur score",pp1)
             if reached_numbers(pp1):
                 print(p1_name,'Won')
                 break
         else:
            print(p2_name,'your turn')
                 
            c=input('press 1 to continue 0 to quit:')
            if(c==0):
                print(p1_name,'scored',pp1)
                print(p2_name,'scored',pp2)
                print('Quitting the game..thanks for playing')
                break
                         
            dice=random.randint(1,6)  
            print('dice showed',dice)
            pp2=pp2+dice
            pp2=check_ladder(pp2)
            pp2=check_snake(pp2)
            if pp2>numbers:
                 pp2=numbers
            print(p2_name,"ur score",pp2)
            if reached_numbers(pp2):
                print(p2_name,'Won')
                break    
         turn=turn+1    
             
          

play()
```
